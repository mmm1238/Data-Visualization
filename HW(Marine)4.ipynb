```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = np.array([
    [30, 20, 50],
    [10, 40, 50],
    [25, 25, 50]
])
row_sums = data.sum(axis=1, keepdims=True)
percent_data = data / row_sums * 100
fig = plt.figure(figsize=(12, 6))
grid = fig.add_gridspec(2, 2)
#Chart1
ax1 = fig.add_subplot(grid[0, 0])
c1 = ax1.imshow(percent_data, cmap='Blues', aspect='auto')
ax1.set_title("chart 1: % per row")
ax1.set_xticks(range(data.shape[1]))
ax1.set_xticklabels(['A', 'B', 'C'])
ax1.set_yticks(range(data.shape[0]))
ax1.set_yticklabels(['Row 1', 'Row 2', 'Row 3'])
fig.colorbar(c1, ax=ax1, fraction=0.046, pad=0.04)
#Chart2
ax2 = fig.add_subplot(grid[0, 1])
c2 = ax2.imshow(data, cmap='Oranges', aspect='auto')
ax2.set_title("chart 2: absolute values")
ax2.set_xticks(range(data.shape[1]))
ax2.set_xticklabels(['A', 'B', 'C'])
ax2.set_yticks(range(data.shape[0]))
ax2.set_yticklabels(['Row 1', 'Row 2', 'Row 3'])
fig.colorbar(c2, ax=ax2, fraction=0.046, pad=0.04)
#Chart3
ax3 = fig.add_subplot(grid[1, :])
ax3.axis('off')
description = (
    "Chart 1 shows the data as percentages where each row sums up to 100%.\n"
    "Chart 2 shows the same data using absolute values.\n"
    "You can observe how the same data appears differently when normalized versus when shown in raw form."
)
ax3.text(0.5, 0.5, description, ha='center', va='center', wrap=True, fontsize=12)

plt.tight_layout()
plt.show()
```
